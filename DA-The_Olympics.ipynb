### Find the number of athletes for each year

```python
# first we will extract the year from the (Year) column of the data as an integer
# then counts the number of unique values in that column for each year
# in the end we will groups the data by the year

df2.groupby(df2['Year'])['Name'].nunique()
```

### Which athlete participated the most? (his/her name and ID)

```python
df2['Name'].nunique()
```

```python
# group the data by athlete and count the number of unique years
# find the athlete with the highest count
# print the name and id of the athlete who participated the most
athlete_count = df2.groupby('Name')['Year'].nunique()
most_participated = athlete_count.idxmax()
print(most_participated)
df2[df2["Name"] == most_participated]["ID"].iloc[0]
```

### Find the total missing data for each year?

```python
df4 = df2.groupby('Year').apply(lambda x: x.isnull().sum())
pd.DataFrame(df4.sum(axis=1), columns=['The total of missing data'])
```

```python
pd.DataFrame(df4.sum(axis=1), columns=['The total of missing data']).plot(kind='bar', width=0.5, figsize=(10, 6))

plt.xticks(rotation=65)
plt.show()
```

```python
# The strongest positive correlation is between Weight and Height, with a value of 0.80.
# This indicates that as the height increases, so does its Weight.

df2_numerical_features = df2.select_dtypes(include='number')
df2_numerical_features.columns
corr = df2_numerical_features.corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', square=True)
plt.title('Heatmap of Numerical Features')
plt.show()
```

### Has the number of athletes changed over time for Summer and Winter?

```python
# Non Detailed first view
# here we can see a first view for the difference in the number
# of athletes that participated in both seasons

df_seasons = df2.groupby('Season')['ID'].nunique()
df_seasons.plot(kind='bar')
plt.show()
```

```python
# Detailed 2nd View
# as we can see below that there is a big difference in the number
# of athletes that participated in both seasons (summer and winter)
# and after the year 1984 there is an increase in the number of participants
# that could be a result of advanced technology or more flexible rules

year_season = df2.groupby(['Year', 'Season'])['ID'].nunique()

# Plot
year_season.plot(kind='bar', width=0.5, figsize=(15, 8))
plt.title('Number of Athletes Participated in Each Season and Year')
plt.xlabel('Year - Season')
plt.ylabel('Number of Athletes')
plt.show()
```
